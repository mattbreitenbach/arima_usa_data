```python
#pip install fredapi
from fredapi import Fred
import pandas as pd
import numpy as np
```

```python
with open("API_KEY.txt") as txt:
    API_KEY = txt.read()

fred = Fred(api_key=API_KEY)
```

```python
inflation_search = fred.search('U.S Inflation CPI', limit=5)
inflation_search.head()
```

```python
income_search = fred.search('Median Household Income in the United State', limit=5)
income_search.head()
```

```python
cpi_us = fred.get_series(series_id='CPIAUCSL', observation_start="1990-01-01" ,observation_end="2020-01-01")
median_income_us = fred.get_series(series_id='MEHOINUSA646N', observation_start="1990-01-01" ,observation_end="2020-01-01")

cpi_us = cpi_us.to_frame(name="cpi_us")
cpi_us.reset_index(inplace=True)
cpi_us.rename( columns={"index" :"date"}, inplace=True )

median_income_us = median_income_us.to_frame(name="median_income_us")
median_income_us.reset_index(inplace=True)
median_income_us.rename( columns={"index" :"date"}, inplace=True )
```

```python
data = pd.merge(cpi_us, median_income_us, how="right", on="date")
data.reset_index(drop=True)
data.head()
```

```python
data["real_mi_us"] = data["median_income_us"] * (data["cpi_us"][len(data["cpi_us"])-1]/data["cpi_us"])
data["per_change_real_mi"] = data["real_mi_us"].pct_change()*100
data.head()
```

```python
data.to_csv(path_or_buf="./data/data.csv", index=False)
```
